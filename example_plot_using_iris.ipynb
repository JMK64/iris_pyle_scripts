```python
import iris
import cartopy
from mpl_toolkits.basemap import Basemap, addcyclic, cm, shiftgrid
import numpy as np
import pylab as plt
import iris
import iris.plot as iplt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
```

```python
data = iris.load_cube('/shared/netscratch/jmk64/monsoon_work/ensemble_mean_2005_2014_o3.nc')
```

```python
# untility cartopy plot
def make_map(bbox):
    fig, ax = plt.subplots(figsize=(20, 20), subplot_kw=dict(projection=ccrs.PlateCarree()))
    ax.set_extent(bbox)
    # bbox format for when i forget  [x0, x1, y0, y1]
    # if masking out land values
    # LAND = cfeature.NaturalEarthFeature('physical', 'land', '50m',
    #                                         edgecolor='face',
    #                                         facecolor=cfeature.COLORS['land']
    #ax.add_feature(LAND, facecolor='white')
    ax.coastlines(resolution='50m')
    gl = ax.gridlines(draw_labels=True, linestyle='--', alpha=0.5, color='gray')
    gl.xlabels_top = gl.ylabels_right = False
    gl.xformatter = LONGITUDE_FORMATTER
    gl.yformatter = LATITUDE_FORMATTER
    return fig, ax
```

```python
plt.clf()
# fix missing values on 0 to 360 grid
data = data.intersection(longitude=(-180, 180))
# get lats and lons
lats = data.coord('latitude').points
lons = data.coord('longitude').points
x, y = lons, lats
# clear figure
plt.clf()
# set bounding box
bbox = [-180, 180, -90, 90]
# make figure axes
fig, ax = make_map(bbox=bbox)
# plot data
cs = ax.pcolormesh(x, y, data.data[0,0,:,:], vmin=0, vmax=10e-8, cmap=plt.cm.Reds)
# add colorbar
cbar = fig.colorbar(cs, orientation='vertical', shrink=0.45, extend='max', pad=0.025)
```
